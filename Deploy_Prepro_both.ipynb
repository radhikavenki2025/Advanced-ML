```python
import pickle
import joblib
```

```python
scaler = joblib.load('scalerx.pkl')
```

```python
preinput=scaler.transform([[1300,12000,4000,0,1]])
```

```python
preinput
```

```python
loaded_model=pickle.load(open("SVMR_finalmodel_both.sav",'rb'))

result=loaded_model.predict(preinput)
```

```python
scalery=joblib.load('scalery.pkl')

preoutput=scalery.inverse_transform(result.reshape(-1,1))
```
